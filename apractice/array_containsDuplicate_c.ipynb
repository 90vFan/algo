```c
#include <stdio.h>
#include <assert.h>
#define bool int
#define false 0
#define true 1

void swap(int *x, int *y) {
    int temp;
    temp = *x;
    *x = *y;
    *y = temp;
}

int partition(int *nums, int low, int high) {
    int pivot;
    int mid = (low + high) / 2;
    swap(&nums[mid], &nums[low]);
    pivot = nums[low];
    
    while (low < high) {
        while (low < high && nums[high] >= pivot) {
            high--;
        }
        nums[low] = nums[high];
        
        while (low < high && nums[low] <= pivot) {
            low++;
        }
        nums[high] = nums[low];
    }
    
    nums[low] = pivot;
    return low;
}

void quick_sort(int *nums, int low, int high) {
    if (low < high) {
        int pivot = partition(nums, low, high);
        quick_sort(nums, low, pivot - 1);
        quick_sort(nums, pivot + 1, high);
    }
}

bool containsDuplicate(int *nums, int numsSize) {
    if (numsSize < 1) return false;
    int flag, i;
    quick_sort(nums, 0, numsSize-1);
    for (i = 0; i < numsSize -1; i++) {
        if (nums[i] == nums[i+1]) return true;
    }
    return false;
}

int main() {
    int inputs[] = {1,2,3,1};
    bool output = containsDuplicate(inputs, 4);
    printf("\nresult: %d\n", output);
    assert(output == true);
    
    int inputs2[] = {1,2,3,4};
    bool output2 = containsDuplicate(inputs2, 4);
    printf("\nresult: %d\n", output2);
    assert(output2 == false);
    
    int inputs3[] = {1,1,1,3,3,4,3,2,4,2};
    bool output3 = containsDuplicate(inputs3, 10);
    printf("\nresult: %d\n", output3);
    assert(output3 == true);
    
    int inputs4[] = {0};
    bool output4 = containsDuplicate(inputs4, 0);
    printf("\nresult: %d\n", output4);
    assert(output4 == false);
}
```
